### Preprocessing and visualizing data

```python
class Loader32(Dataset):
    
    def __init__(self, data, length):
        assert len(data) >= length
        self.data = data
        self.length = length
    
    def __getitem__(self, idx):
        return torch.tensor(self.data[idx:idx+self.length]).reshape(-1, self.length).to(torch.float32)
        
    def __len__(self):
        return max(len(self.data)-self.length, 0)

returns = data.shift(1)/data - 1
log_returns = np.log(data/data.shift(1))[1:].to_numpy().reshape(-1, 3)
standardScaler1 = StandardScaler()
standardScaler2 = StandardScaler()
gaussianize = Gaussianize()
#try other scaler methods (minmax)
#remove second scaler

log_returns_preprocessed = gaussianize.fit_transform(standardScaler1.fit_transform(log_returns))
receptive_field_size = 127  # p. 17
data_size = log_returns.shape
print(data_size)
```

# Training


```python
#Todo: wrap conditional statement around conditional noise to make sure idx>context length
import torch.optim as optim
from tqdm import tqdm
generator = Generator().to(device)

discloss = []
genloss = []

train = True

if train:
    context_len = 100
    discriminator = Discriminator((seq_len)).to(device)

    disc_optimizer = optim.RMSprop(discriminator.parameters(), lr=lr)
    gen_optimizer = optim.RMSprop(generator.parameters(), lr=lr)

    dataset = Loader32(log_returns_preprocessed, seq_len)
    dataloader = torch.utils.data.DataLoader(dataset, batch_size=batch_size)

    #separate contextset which loads returns of context_len
    contextset = Loader32(log_returns_preprocessed,  context_len)

    t = tqdm(range(num_epochs))
    for epoch in t:
        for idx, data in enumerate(dataloader, context_len):
            discriminator.zero_grad()
            real = data.to(device).reshape(-1, seq_len)
            #context = tensor containing data[index - context_length: index] of length context_len hence why contextset is used
            context = contextset[idx-context_len]
            noise = torch.randn((1, seq_len), device=device)
            
            #concatenate context and noise, context and real
            cond_noise = torch.cat((context, noise), axis=0)

            fake = generator(cond_noise).detach()

            disc_loss = -torch.mean(discriminator(real)) + torch.mean(discriminator(fake))
            disc_loss.backward()
            disc_optimizer.step()
            discloss.append(disc_loss.item())

            for dp in discriminator.parameters():
                dp.data.clamp_(-clip, clip)
    
            if idx % 5 == 0: #2-5
                generator.zero_grad()
                gen_loss = -torch.mean(discriminator(generator(cond_noise)))    
                gen_loss.backward()
                gen_optimizer.step() 
                genloss.append(gen_loss.item())           
        t.set_description('Discriminator Loss: %.8f Generator Loss: %.8f' % (disc_loss.item(), gen_loss.item()))
            
    # Save
    torch.save(generator, f'SPYAAPLTLT/test1.pth')
```

### Generating Synthetic Samples

```python

dataloader = Loader32(log_returns_preprocessed, seq_len)
contextset = Loader32(log_returns_preprocessed,  context_len)
context = contextset[indx-context_len]
noise = torch.randn((1, seq_len), device=device)

cond_noise = torch.cat((context, noise))

generator = torch.load("SPYAAPLTLT/test1.pth")
y = generator(cond_noise).cpu().detach().squeeze()

y_reshaped = y.reshape(-1, 3)
y_reshaped = (y_reshaped - y_reshaped.mean(axis=0)) / y_reshaped.std(axis=0)

# Step 1: Transpose the array to work with columns
y_transposed = y_reshaped.T

# Step 2: Initialize a list to store the transformed columns
transformed_columns = []

# Step 3: Process each column individually
for column in y_transposed:
    # Reshape the column to a 2D array (required by the transform function)
    reshaped_column = column.reshape(-1, 1).numpy()
    # Apply the inverse Gaussian transformation
    transformed_column = gaussianize.inverse_transform(reshaped_column)
   
    # Flatten the column vector back to a 1D array
    flattened_column = transformed_column.flatten()
    # Append the transformed column to the list
    transformed_columns.append(flattened_column)
   
# Step 4: Convert the list of transformed columns back to a 2D numpy array
y_transformed = np.array(transformed_columns)

# Step 5: Transpose the array back to its original orientation
y_reshaped = y_transformed.T
y_reshaped = standardScaler1.inverse_transform(y_reshaped)
y = y_reshaped.reshape(3, -1)

print(y.shape)

fig, ax = plt.subplots(figsize=(12, 6))
print(y.shape)

ax.plot(np.cumsum(y[0, :]), alpha=0.75, label=f"AAPL")
ax.plot(np.cumsum(y[1, :]), alpha=0.75, label=f"MSFT")

ax.set_title('Generated log return paths')
ax.set_xlabel('days')
ax.set_ylabel('Cumulative log return')
ax.legend()

plt.show()
```
